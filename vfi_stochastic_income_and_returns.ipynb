```python
# Value Function Iteration with IID Income
# Greg Kaplan 2017
# Translated by Tom Sweeney Dec 2020

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve 
from discrete_normal import discrete_normal
from itertools import chain
import pandas as pd
```

```python
# PARAMETERS

## preferences
risk_aver = 2
beta = 0.971

## return risk: discretized N(mu,sigma^2)
mu_r = 0.03
sd_r = 0.003
nr = 5

## income risk: discretized N(mu,sigma^2)
mu_y = 1
sd_y = 0.2
ny = 5

## asset grids
na = 4000
amax = 1000
borrow_lim = 0
agrid_par = 1 # 1 for linear, 0 for L-shaped

## computation
max_iter = 1000
tol_iter = 1.0e-1
Nsim = 50000 #specify number of individuals in economy
Tsim = 1000 #specify number of time periods to simulate for
```

```python
# OPTIONS
Display = 1
DoSimulate = 1
MakePlots = 1

# DRAW RANDOM NUMBERS
np.random.seed(2020)


# SET UP GRIDS

## assets
agrid = np.linspace(0,1,na).reshape(na,1)
agrid = agrid**(1/agrid_par)
agrid = borrow_lim + (amax-borrow_lim)*agrid

## income: disretize normal distribution
width_y = fsolve(lambda x: discrete_normal(ny,mu_y,sd_y,x)[0],2)
temp, ygrid, ydist = discrete_normal(ny,mu_y,sd_y,width_y)
ycumdist = np.cumsum(ydist)

## returns: disretize normal distribution
width_r = fsolve(lambda x: discrete_normal(nr,mu_r,sd_r,x)[0],2)
temp, rgrid, rdist = discrete_normal(nr,mu_r,sd_r,width_r)
rcumdist = np.cumsum(rdist)

# Set up Rgrid = 1+rgrid
Rgrid = 1+rgrid

# UTILITY FUNCTION
if risk_aver==1:
    u = lambda c: np.log(c)
else:
    u = lambda c: (c**(1-risk_aver)-1)/(1-risk_aver) 
```

```python
# INITIALIZE VALUE FUNCTION

Vguess = np.zeros((nr,na,ny))
for ir in range(0,nr):
    for iy in range(0,ny):
        Vguess[ir,:,iy] = u(rgrid[ir]*agrid[0]+ygrid[iy])/(1-beta)
        
```

```python
# ITERATE ON VALUE FUNCTION

V_full=Vguess.copy()
sav_full = np.zeros((nr,na,ny))
savind_full = np.zeros((nr,na,ny), dtype=int)
con_full = np.zeros((nr,na,ny))


for ir in range(0,nr):

    V = Vguess[ir].copy()

    Vdiff = 1
    Iter = 0
    
    if Display >= 1:
        print('Return index: ' + str(ir), ' Return value: ' + str(Rgrid[ir]))

    while Iter <= max_iter and Vdiff > tol_iter:
        Iter = Iter + 1
        Vlast = V.copy()
        V = np.zeros((na,ny))
        sav = np.zeros((na,ny))
        savind = np.zeros((na,ny), dtype=int)
        con = np.zeros((na,ny))

        ## loop over assets
        for ia in range(0,na):

            ## loop over income
            for iy in range(0,ny):
                cash = Rgrid[ir]*agrid[ia] + ygrid[iy]
                Vchoice = u(np.maximum(cash-agrid,1.0e-10)) + beta*(Vlast @ ydist)           
                V[ia,iy] = np.max(Vchoice)
                savind[ia,iy] = np.argmax(Vchoice)
                sav[ia,iy] = agrid[savind[ia,iy]]
                con[ia,iy] = cash - sav[ia,iy]

        Vdiff = np.max(abs(V-Vlast))
        if Display >= 1:
            print('Iteration no. ' + str(Iter), ' max val fn diff is ' + str(Vdiff))

    V_full[ir] = V
    sav_full[ir] = sav
    savind_full[ir] = savind
    con_full[ir] = con
    
    if Display >= 1:
        print('Return index ' + str(ir), ' Finished.')
        print('-------------------------------------')
```

```python
## Assume return risk is systemic. Create array of size (1*Tsim) realized returns at period t

# 'rdist' is the probability of each realization
##Turn rdist values into a list------##
rlist = rdist.tolist()

r_probabilities = []
for sublist in rlist:
    for x in sublist:
        r_probabilities.append(x)
##-----------------------------------##

rindsim = np.zeros((1,Tsim),dtype=int)
for t in range(0,Tsim):
    rindsim[0][t]=np.random.choice([0,1,2,3,4], p=r_probabilities) #randomly choose path of economy returns

## Check that returns are normally distributed
import seaborn as sns    
sns.distplot(pd.DataFrame(rindsim.transpose()),kde=False)
```

```python
# SIMULATE
if DoSimulate == 1:
    
    ## Create a matrix of size (Nsim * Tsim) of income realizations for Nsim individuals over Tsim periods
    yrand = np.random.rand(Nsim,Tsim) # draw random realizations from U~[0,1)
    
    ## Create matrices of size (Nsim * Tsim)
    yindsim = np.zeros((Nsim,Tsim), dtype=int) #this is a matrix of every individuals' income at period t
    aindsim = np.zeros((Nsim,Tsim), dtype=int) #this is a matrix of every individuals' assets at period t
    
    ## initial assets
    aindsim[:,0] = 0 #selecting a vector of size (Nsim * 1); this command sets the first period assets to 0 for every individual
    
    
    ##-------------------------------------------------------------------------------------------##
    ##This part is all to set randomized income realizations according to the discretized normal realizations we defined earlier
    
    ## loop over time periods
    for it in range(0,Tsim):
        if Display >= 1 and (it+1)%100 == 0:
            print(' Simulating, time period ' + str(it+1))

        #this below operation sets the 'minimum' income realizations of yindsim to 0
        """Code explanation:
            (1) yrand[:,it] selects vector of size (Nsim*1) individuals' income at period t that we are iterating on
            (2) check if yrand[:,it] is weakly smaller than the minimum value of ycumdist
            (3) select all rows of the vector 'yindsim at time t' that are less than the minimum of (2), and sets all these rows to 0
        """        
        yindsim[yrand[:,it]<=ycumdist[0],it] = 0 
            
        #following a similar logic, we set the randomized income realizations to the discretized normal income realization values
        #Now, the yindsim vector at time t only has income values that are in ycumdist            
        for iy in range(1,ny):
            yindsim[np.logical_and(yrand[:,it]>ycumdist[iy-1], yrand[:,it]<=ycumdist[iy]),it] = iy

        ##-------------------------------------------------------------------------------------------##


        """--Now we can simulate the path of assets--"""        

        ## asset choice
        """Code explanation:
            (1) yrand[:,it] selects vector of size (Nsim*1) individuals' income at period t that we are iterating on
            (2) Select only the rows where yindsim[:,it]==iy. We loop over every iy in ny, so every row will be selected at some point
            (3) Individuals will choose optimal assets of period t+1 according to their savings policy function given economy interest rate
        """
        if it < Tsim-1:
            for iy in range(0,ny):
                aindsim[yindsim[:,it]==iy,it+1] = savind_full[rindsim[0][it],aindsim[yindsim[:,it]==iy,it],iy]

    ## assign actual asset and income values
    asim = agrid[aindsim]
    ysim = ygrid[yindsim]
```

```python
# MAKE PLOTS
if MakePlots==1:
    
    ## consumption policy function
    plt.plot(agrid,con[:,0],'b-',label = 'Lowest income state')
    plt.plot(agrid,con[:,ny-1],'r-', label = 'Highest income state')
    plt.grid()
    plt.xlim((0,amax))
    ### plt.title('Consumption Policy Function')
    plt.title('Consumption')
    plt.legend()
    plt.show()

    ## savings policy function
    plt.plot(agrid,sav[:,0]-agrid[:,0],'b-')
    plt.plot(agrid,sav[:,ny-1]-agrid[:,0],'r-')
    plt.plot(agrid,np.zeros((na,1)),'k',linewidth=0.5)
    plt.grid()
    plt.xlim((0,amax))
    ### plt.title('Savings Policy Function (a''-a)')
    plt.title('Savings')
    plt.show()
    
    ## nice zoom
    xlimits = (0,1)
    xlimind = np.ones(na, dtype=bool)
    if np.min(agrid) < xlimits[0]:
        xlimind = np.logical_and(xlimind,(agrid[:,0]>=np.max(agrid[agrid<xlimits[0]])))
    elif np.min(agrid) > xlimits[1]:
        xlimind = 0
    if np.max(agrid) > xlimits[1]:
        xlimind = np.logical_and(xlimind,(agrid[:,0]<=np.min(agrid[agrid>xlimits[1]])))
    elif np.max(agrid) < xlimits[0]:
        xlimind = 0

    ## consumption policy function: zoomed in
    plt.plot(agrid[xlimind],con[xlimind,0],'b-o',linewidth=2)
    plt.plot(agrid[xlimind],con[xlimind,ny-1],'r-o',linewidth=2)
    plt.grid()
    plt.xlim(xlimits)
    plt.title('Consumption: Zoomed')
    plt.show()

    ## savings policy function: zoomed in
    plt.plot(agrid[xlimind],sav[xlimind,0]-agrid[xlimind,0],'b-o',linewidth=2)
    plt.plot(agrid[xlimind],sav[xlimind,ny-1]-agrid[xlimind,0],'r-o',linewidth=2)
    plt.plot(agrid,np.zeros((na,1)),'k',linewidth =0.5)
    plt.grid()
    plt.xlim(xlimits)
    plt.title('Savings: Zoomed (a\'-a)')
    plt.show()

    ## income distribution
    plt.hist(ysim[:,Tsim-1],len(ygrid),facecolor=(0,0.5,0.5),edgecolor='blue')
    plt.ylabel('')
    plt.title('Income distribution')
    plt.show()

    ## asset distribution
    plt.hist(asim[:,Tsim-1],40,facecolor=(.7,.7,.7),edgecolor='black')
    plt.ylabel('')
    plt.title('Asset distribution')
    plt.show()

    ## convergence check
    plt.plot(range(0,Tsim),np.mean(asim,0),'k-',linewidth=1.5)
    plt.xlabel('Time Period')
    plt.title('Mean Asset Convergence')
    plt.show()

    ## asset distribution statistics
    aysim = asim[:,Tsim-1]/np.mean(ysim[:,Tsim-1])
    print('Mean assets: ' + str(np.mean(aysim)))
    print('Fraction borrowing constrained: ' + str(np.sum(aysim==borrow_lim)/Nsim * 100) + '%')
    print('10th Percentile: ' + str(np.quantile(aysim,.1)))
    print('50th Percentile: ' + str(np.quantile(aysim,.5)))
    print('90th Percentile: ' + str(np.quantile(aysim,.9)))
    print('99th Percentile: ' + str(np.quantile(aysim,.99)))
```
